```python
df.columns
```

```python
df.rename(columns={'Marital_Status':'Married'},inplace=True)
```

```python
def chang(x):
    if x==0:
        return 'N'
    else:
        return 'Y'
```

```python
df['Married']=df['Married'].transform(chang)
```

```python
df
```

```python
def categorize_age(x):
    if x<=17:
        return '0-17'
    elif x<=25:
        return '18-25'
    if x<=35:
        return '26-35'
    elif x<=45:
        return '36-45'
    if x<=50:
        return '46-50'
    elif x<=55:
        return '51-55'
    else:
        return '55+'
```

```python
df['Age_Group']=df['Age'].apply(categorize_age)
```

# Exploratory Data Analysis


## Gender wise 

```python
ax=sns.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
amount_by_gender=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
d=sns.barplot(x='Gender',y='Amount',data=amount_by_gender)
# for bars in d.containers:
#     d.bar_label(bars)
```

From the above two graphs we can clearly see that majority of the customers are females ans also majority of the revenue comes from them 

## AgeWise

```python
ax=sns.countplot(data=df,x='Age_Group',hue='Gender')
```

```python
age_wise=df.groupby(['Age_Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
# df.head()
```

```python
# hue=['0-17','18-25','26-35','36-45','46-50','51-55','55+']
# f=pd.DataFrame(age_wise)
age_wise
```

```python
sns.barplot(x='Age_Group',y='Amount',data=age_wise)
```

From the above two graphs we can conclude that the age group 26-35 are the largest buyers as well as they have the largest purchasing power.Particularly females make up a very large portion of it. 

## State Wise
